```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('datasets/IllinoisStudentsByCurriculum.csv',dtype=str)
df = df.fillna('0')

def handle_x(x):
    try:
        return np.int32(x.replace(",", ""))
    except:
        return 0

df['Total'] = df['Total'].apply(lambda x: handle_x(x))
```
