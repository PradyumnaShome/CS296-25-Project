```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('datasets/uiuc-students-by-curriculum-cleaned.csv',dtype=str)
```

```python
df.head()
```

```python
new_df = df[df['Fall'].isin(['1985', '2018'])]
```

```python
new_df.head()
```

```python
major_df = new_df.loc[:, new_df.columns.intersection(['Fall', 'Total', 'Major Name'])]
major_df['Total'] = major_df['Total'].astype(np.int32)
```

```python
major_df = major_df.drop(major_df[major_df['Total'] < 100].index)
```

```python
set(major_df['Fall'].values)
```

```python
majors_1985 = major_df[major_df['Fall'] == '1985']
majors_1985 = majors_1985.groupby(by='Major Name',as_index=False).max()

majors_2018 = major_df[major_df['Fall'] == '2018']
majors_2018 = majors_2018.groupby(by='Major Name',as_index=False).max()
```

```python
for_slopegraph_df = pd.merge(majors_1985, majors_2018, how='inner',on='Major Name')
```

```python
for_slopegraph_df.head()
```

```python
for_slopegraph_df.rename(columns={'Total_x':'Total_1985','Total_y':'Total_2018'}, inplace=True)
```

```python
for_slopegraph_df.head()
```

```python
for_slopegraph_df.drop(['Fall_x', 'Fall_y'], axis='columns',inplace=True)
```

```python
for_slopegraph_df['Percent_change'] = for_slopegraph_df['Total_2018'] / for_slopegraph_df['Total_1985']
```

```python
for_slopegraph_df.to_csv('change_in_major.csv')
```
