```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('datasets/uiuc-students-by-curriculum-cleaned.csv',dtype=str)
```

```python
df.head()
```

```python
major_df = df.loc[:, df.columns.intersection(['Fall', 'Total', 'Major Name', 'College', 'Male', 'Female', 'Illinois', 'Non-Illinois'])]
major_df['Total'] = major_df['Total'].astype(np.int32)
```

```python
# major_df = major_df.drop(major_df[major_df['Total'] < 100].index)

grouped_by_major = list(major_df.groupby(by='Major Name',as_index=False))
```

```python
major_df = major_df.groupby(by=['Fall', 'Major Name'],as_index=False).max()
```

```python
major_df[major_df['Major Name'] == 'Computer Science']
```

```python
major_df = major_df.sort_values(by=['College','Major Name'])
```

```python
major_df = major_df.pivot(index='Major Name',columns='Fall')
major_df = major_df.dropna(thresh=20)
major_df.head()
```

```python
major_df.columns = ['_'.join(col).strip() for col in major_df.columns.values]
major_df.columns
```

```python
# major_df = major_df.reset_index(inplace=True)
```

```python
major_df = major_df.fillna(0)
major_df["College"] = major_df["College_2018"]
major_df = major_df.drop(columns=["College_" + str(x) for x in range(1980, 2019)])
```
