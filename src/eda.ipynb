```python
import pandas as pd
import numpy as numpy

df = pd.read_csv('datasets/IllinoisStudentsByCurriculum.csv',low_memory=False)
df.fillna(0, inplace=True)
```

```python
sum_over = ['Total', 'African American', 'Asian American', 'Caucasian', 'Female','Hawaiian/Pacific Isl','Hispanic','Illinois', 'International','Male','Multiracial','Native American','Non-Illinois','Unknown','Unknown.1']

for col in sum_over:
    if df[col].dtype == str or df[col].dtype == object:
        df[col] = df[col].str.replace(',','')

df[sum_over] = df[sum_over].astype(np.int64, errors='ignore')

# df['Total'].dtype
grouped_by_major = df.groupby('Major Name',as_index=False)

grouped_by_major.get_group('Computer Science').columns

# major_dfs = []

# for name, group in grouped_by_major:
#     group = group.groupby('Fall',as_index=False).agg(sum)

# new_df = pd.concat(major_dfs)
# new_df['Major Name']
```

```python
grouped_by_major = df.groupby('Major Name')
```

```python
cs_df = grouped_by_major.get_group('Computer Science').copy()

sum_over = ['Total', 'African American', 'Asian American', 'Caucasian', 'Female','Hawaiian/Pacific Isl','Hispanic','Illinois', 'International','Male','Multiracial','Native American','Non-Illinois','Unknown','Unknown.1']

cs_df[sum_over] = cs_df[sum_over].astype(np.int64)

cs_df = cs_df.groupby('Fall',as_index=False).agg(sum)
```

```python
cs_df.head()
```

```python
cs_df['Percent Female'] = cs_df['Female']/cs_df['Total']

cs_df.plot.scatter(x='Fall',y='Percent Female',title='Percent Female in Computer Science')
```
