```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/IllinoisStudentsByCurriculum.csv', low_memory=False, dtype={'Male':str, 'Female':str})
```

```python
df.head()
```

```python
df = df.fillna('0')
```

```python
def handle_x(x):
    try:
        return np.int32(x.replace(",", ""))
    except:
        return 0
        
df['Male'] = df['Male'].apply(lambda x: handle_x(x))
df['Female'] = df['Female'].apply(lambda x: handle_x(x))
```
