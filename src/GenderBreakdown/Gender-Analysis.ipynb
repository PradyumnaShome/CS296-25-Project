```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/IllinoisStudentsByCurriculum.csv', low_memory=False, dtype={'College': str, 'Total':str, 'Male':str, 'Female':str, 'Illinois': str, 'Non-Illinois': str})
```

```python
df.head()
```

```python
df.dropna(axis=0, subset=['College', 'Major Name', 'Total'], inplace=True)
df = df.fillna('0')
```

```python
def handle_x(x):
    if (type(x) == np.int32):
        return x
    try:
        return np.int32(x.replace(",", ""))
    except:
        return 0

df['Total'] = df['Total'].apply(handle_x)
df['Male'] = df['Male'].apply(handle_x)
df['Female'] = df['Female'].apply(handle_x)
df['Illinois'] = df['Illinois'].apply(handle_x)
df['Non-Illinois'] = df['Non-Illinois'].apply(handle_x)
```

```python
df.head()
```

```python
df_gender_sum = df.groupby(['Fall', 'College', 'Major Name']).sum().reset_index()
```

```python
df_grouped = df_gender_sum.groupby(['College', 'Fall'])
```

```python
for info, group in df_grouped:
    year = info[1]
    college = info[0]
    name = 'datasets/' + str(year) + '-' + college + '-GB.csv'
    group = group.nlargest(10, 'Total')
    group.to_csv(name)
```
